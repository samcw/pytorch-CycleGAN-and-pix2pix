```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import os
import random
from matplotlib import rcParams

config = {
    "font.size": 13,
    'font.sans-serif': "Times New Roman"
}

rcParams.update(config)
```

```python
FLOPs_MobileNet=[9, 21, 42]
FLOPs_ShuffleNet=[10, 23, 44]
FLOPs_GhostNet=[8, 20, 36]
FLOPs_SGhostNet=[7, 19, 34]

Acc_MobileNet=[88.1, 90.5, 92.4]
Acc_ShuffleNet=[88.4, 90.9, 92.2]
Acc_GhostNet=[89.2, 91.4, 93.2]
Acc_SGhostNet=[90.0, 92.0, 93.6]

plt.plot(FLOPs_MobileNet, Acc_MobileNet, linewidth=1, linestyle="--", color='b', label='MobileNetV2', marker=">")
plt.plot(FLOPs_ShuffleNet, Acc_ShuffleNet, linewidth=1, linestyle="--", color='r', label='ShuffleNetV2', marker="<")
plt.plot(FLOPs_GhostNet, Acc_GhostNet, linewidth=1, linestyle="--", color='g', label='GhostNet', marker="^")
plt.plot(FLOPs_SGhostNet, Acc_SGhostNet, linewidth=1, color='y', label='S-GhostNet', marker="v")
plt.xlabel('FLOPs (M)')
plt.ylabel('Top-5 Acc (%)')
plt.title('')
plt.legend()

plt.savefig('FLOPS.png', dpi=256)
```
